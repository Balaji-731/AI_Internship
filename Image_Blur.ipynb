```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
## open the image
im=Image.open("/content/download.jpg").convert("RGB")
## create a solid color image(same size)
blue_tint=Image.new("RGB",im.size,(0,0,255))
## blend the original with the blue tint
tinted=Image.blend(im, blue_tint,alpha=0.5)
## display the image
plt.imshow(tinted)
plt.axis("off")
plt.title("Blue_Tinted Image")
plt.show()
```

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
## open the image
im=Image.open("/content/download.jpg").convert("RGB")
## create a solid color image(same size)
blue_tint=Image.new("RGB",im.size,(0,0,255))
## blend the original with the blue tint
tinted=Image.blend(im, blue_tint,alpha=0.5)
## display the image
plt.imshow(tinted)
plt.axis("on")
plt.title("Blue_Tinted Image")
plt.show()
```

Image Blurring

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# open image
im = Image.open("/content/download.jpg").convert("RGB")

# apply gaussian blur
blurred = im.filter(ImageFilter.GaussianBlur(radius=2))

# display
plt.imshow(blurred)
plt.axis("off")
plt.title("Gaussian Blurred Image")
plt.show()
```

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# open image
im = Image.open("/content/download.jpg").convert("RGB")

# apply blox blur
blurred = im.filter(ImageFilter.BoxBlur(4))
# display
plt.imshow(blurred)
plt.axis("off")
plt.title("Blox Blurred Image")
plt.show()
```

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# open image
im = Image.open("/content/download.jpg").convert("RGB")

# apply normal blur
blurred = im.filter(ImageFilter.BLUR)
# display
plt.imshow(blurred)
plt.axis("off")
plt.title("Blox Blurred Image")
plt.show()
```
