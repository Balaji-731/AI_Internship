```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("/content/Salary_Data.csv")
print(dataset.head())
```

```python
dataset.isnull().any()
```

```python
dataset["Years of Experience"]=dataset["Years of Experience"].fillna(dataset["Years of Experience"].mean())
dataset["Salary"]=dataset["Salary"].fillna(dataset["Salary"].mean())
```

```python
dataset[dataset["Years of Experience"].isnull()]
```

```python
x=pd.DataFrame(dataset.iloc[:,4])
y=dataset.iloc[:,5]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train,color="green")
plt.plot(x_train,regressor.predict(x_train),color="red")
plt.title("Salary vs Experience(Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test,y_test,color="green")
plt.plot(x_test,regressor.predict(x_test),color="red")
plt.title("Salary vs Experience(Testing Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
